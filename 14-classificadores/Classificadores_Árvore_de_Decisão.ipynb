```python
# importing libraries
import numpy as nm
import matplotlib.pyplot as mtp
import pandas as pd

#importing datasets
data_set = pd.read_csv('https://raw.githubusercontent.com/celsocrivelaro/simple-datasets/main/user_data.csv')
data_set
```

```python
feature_names = ["Idade", "Salário"]
target_names=["Comprador", "Não comprador"]
```

```python
#Extracting Independent and dependent Variable
x = data_set.iloc[:, [2,3]].values
y = data_set.iloc[:, 4].values
```

```python
# Splitting the dataset into training and test set.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
# Treinamento do modelo
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy', random_state=0, max_depth=5)
classifier.fit(x_train, y_train)
```

```python
a = classifier.predict(nm.array([[30, 87000]]))
a
```

# Visualizando o modelo


```python
from sklearn import tree
tree.plot_tree(classifier)
import matplotlib.pyplot as plt

plt.figure(figsize=(16,16))  # set plot size (denoted in inches)
tree.plot_tree(classifier, fontsize=10, feature_names=feature_names, class_names=target_names)
plt.show()
```
